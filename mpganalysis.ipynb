```python
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set_palette("bright")
```

```python
mpg = sns.load_dataset('mpg')
mpg
```

```python
#Create a visual showing the relationship between horsepower and fuel economy.
sns.relplot(data=mpg,x='horsepower',y='mpg')
```

### The scatterplot shows that the more horsepower the car has the lower its fuel economy.

```python
#Create a table summarizing fuel economy by country of origin
```

```python
mpg_per_origin = mpg.groupby('origin').agg({'mpg':'mean', 'cylinders':'mean'})
mpg_per_origin
```

#### Overall the USA makes cars with bigger cylinders but have lower fuel economy on average. Japan has the highest average fuel economy with smaller cylinders.

```python
#Create a visual showing how fuel economy has changed over time
```

```python
sns.relplot(data=mpg, x="model_year", y="mpg", hue="horsepower")
```

### We see an upward trend on fuel economy the older the car year, but it depends on the car's horsepower as well. Older cars with more horsepower had the worse fuel economy.

```python
#Create two other visuals or tables related to fuel economy.
```

```python
sns.relplot(data=mpg, x="model_year", y="mpg", kind="line")
sns.relplot(data=mpg, x="model_year", y="horsepower", kind="line")
```
